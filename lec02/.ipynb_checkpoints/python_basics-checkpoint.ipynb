**Задание 2**. Дано значение объема файла в байтах (*в мегабайте 2^20 байт*). Напишите перевод этого значения в мегабайты в формате: "*Объем файла равен 213.68Mb*".

```python
dim_bytes = int(input("введите объем файла в байтах "))
dim_mb = dim_bytes / 2**20
print ("Объем файла равен ", round(dim_mb,2), "МБ")
```

**Задание 3**. Разработать приложение для определения знака зодиака по дате рождения. Пример:

Введите месяц: март

Введите число: 6

Вывод:
Рыбы

```python
print("Введите дату своего рождения, месяц русскими строчными буквами, дату числом")
month = input("месяц рождения ")
number = int(input ("число "))

if (number < 0 or number > 31):
    print ("Ваш ввод не для гороскопа")
elif (month == "январь" and number < 21) or (month == "декабрь" and number >= 23):
    print ("Козерог")
elif month == "январь" or (month == "февраль" and number < 20):
    print ("Водолей")
elif month == "февраль" or (month == "март" and number < 21):
    print ("Рыбы")
elif month == "март" or (month == "апрель" and number < 21):
    print ("Овен")
elif month == "апрель" or (month == "май" and number < 22):
    print ("Телец")
elif month == "май" or (month == "июнь" and number < 22):
    print ("Близнецы")
elif month == "июнь" or (month == "июль" and number < 23):
    print ("Рак")
elif month == "июль" or (month == "август" and number < 22):
    print ("Лев")
elif month == "август" or (month == "сентябрь" and number < 24):
    print ("Дева")
elif month == "сентябрь" or (month == "октябрь" and number < 24):
    print ("Весы")
elif month == "октябрь" or (month == "ноябрь" and number < 23):
    print ("Скорпион")
elif month == "ноябрь" or month == "декабрь":
    print ("Стрелец")
else: 
    print ("Змееносец")

print ("Конец программы")
```

**Задание 4**. Нужно разработать приложение для финансового планирования.
Приложение учитывает сколько уходит на ипотеку, "на жизнь" и сколько нужно отложить на пенсию.
Пользователь вводит:
- заработанную плату в месяц.
- сколько процентов от ЗП уходит на ипотеку.
- сколько процентов от ЗП уходит "на жизнь".
- сколько раз приходит премия в год.

Остальная часть заработанной платы откладывается на пенсию.

Также пользователю приходит премия в размере зарплаты, от которой половина уходит на отпуск, а вторая половина откладывается.

Программа должна учитывать сколько премий было в год.

Нужно вывести сколько денег тратит пользователь на ипотеку и сколько он накопит за год.

Пример:

Введите заработанную плату в месяц: 100000

Введите сколько процентов уходит на ипотеку: 30

Введите сколько процентов уходит на жизнь: 50

Введите количество премий за год: 2

Вывод:
На ипотеку было потрачено: 360000 рублей
Было накоплено: 340000 рублей

```python
# Надеюсь, что зарплата не бюджетная и круглая
salary = int(input("Введите заработанную плату в месяц: "))
mortgage_cent = float(input("Введите сколько процентов уходит на ипотеку: "))
life_cent = float(input("Введите сколько процентов уходит на жизнь: "))
bonus = int(input("Введите количество премий за год: "))

mortgage = int(12 * salary * mortgage_cent / 100)
saving = int(salary * (0.12 * (100 - mortgage_cent - life_cent) + bonus/2))

print("На ипотеку было потрачено ", mortgage, "рублей. Было накоплено ", saving, "рублей.")
```
