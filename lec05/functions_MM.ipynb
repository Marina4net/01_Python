Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

Перечень полок, на которых находятся документы хранится в следующем виде:

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
def main():
    '''
    Команда для работы с документами
    '''
    while True:
        user_command = input('введите команду для действия(латиница), ? для помощи ').lower()
        if user_command == 'q':
            print('конец работы')
            break
        elif user_command == '?':
            print('Команда для работы с документами:')
            print('p – people – определение владельца документа по номеру документа;')
            print('s – shelf – опредление полки документа по номеру документа;')
            print('l – list – вывод всех имеющихся документов и их владельцев;')
            print('as – add shelf – добавление новой пустой полки;')
            print('d – delete – удаление документа и его атрибутов по номеру документа;')
            print('m – move – перемещение документа с полки на полку;')
            print('a – add – добавление документа и его атрибутов;')
            print('q - завершение работы.')
        elif user_command == 'p':
            people()
        elif user_command == 's':
            shelf()
        elif user_command == 'l':
            list_doc()
        elif user_command == 'as':
            add_shelf()
        elif user_command == 'd':
            del_doc()
        elif user_command == 'm':
            move_doc()
        elif user_command == 'a':
            add_doc()
        else:
            print('неизвестная команда, попробуйте еще')

def check_doc(doc_number):
    '''Проверка существования документа, вспомогательная команда'''
    for diction in documents:
        if diction['number'] == doc_number:
            return True
    return False

def check_shelf(shelf_number):
    '''Проверка существования полки, вспомогательная команда'''
    if shelf_number in directories:
        return True
    return False

def people():
    '''Команда people, ищет владельца документа по номеру документа'''
    doc_number = input("Введите номер документа ")

    if check_doc(doc_number):
        for diction in documents:
            if diction['number'] == doc_number:
                print(f"{diction['name']} владелец документа {doc_number}")
    else: 
        print("По вашему запросу ничего не найдено")

def shelf():
    '''Команда shelf, ищет полку по номеру документа'''

    doc_number = input("Введите номер документа ")
    
    if check_doc(doc_number):
        for key in directories:
            if doc_number in directories[key]:
                print(f"{doc_number} лежит на полке номер {key}")
    else: 
        print("По вашему запросу ничего не найдено")

def list_doc():
    '''Вывод списка всех документов'''

    for diction in documents:
        print('\033[1m', f"{diction['type']} \"{diction['number']}\" \"{diction['name']}\"", '\033[0m')

def add_shelf():
    '''Добавление новой пустой полки'''
    new_shelf = input("введите номер новой полки ")
    
    if check_shelf(new_shelf):
        print(f"полка {new_shelf} существует, повезет в другой раз")
    else:
        directories[new_shelf] = []
        print(f"Полка {new_shelf} добавлена успешно")

def del_doc():
    '''бесследно удаляет документ'''
    
    doc_number = input('Введите номер документа для удаления ')
    if check_doc(doc_number):
        for diction in documents:
            if diction['number'] == doc_number:
                documents.remove(diction)
        for key in directories:
            if doc_number in directories[key]:
                directories[key].remove(doc_number)
    else: 
        print("Такой документ не существует")

def add_doc():
    '''добавление документа'''
    doc_number = input('Введите номер документа для добавления ')
    
    if check_doc(doc_number):
        print('такой документ уже существует')
    else:
        new_user={}
        new_user['number'] = doc_number
        new_user['type'] = input('Введите тип документа ').lower()
        new_user['name'] = input('Введите имя и фамилию владельца').title()
        shelf_number = input('Введите номер полки, куда положить документ ')
        if check_shelf(shelf_number):
            directories[shelf_number].append(doc_number)
            documents.append(new_user)
        else:
            user_choice = input('Полки не существует, добавить полку (у) или потерять все данные (n)?').lower()
            if user_choice == 'y':
                directories[shelf_number] = []
                directories[shelf_number].append(doc_number)
                documents.append(new_user)

def move_doc():
    '''Перемещение документа с полки на полку'''
        
    doc_number = input("Введите номер документа ")
    key_shelf=0
    if check_doc(doc_number):
        for key in directories:
            if doc_number in directories[key]:
                print(f"{doc_number} лежит на полке номер {key}")
                key_shelf = key
        shelf_number = input('Введите номер полки, куда положить документ ')
        if check_shelf(shelf_number):
            directories[key_shelf].remove(doc_number)
            directories[shelf_number].append(doc_number)
        else:
            user_choice = input('Полки не существует, добавить полку (у) или потерять все данные (n)?').lower()
            if user_choice == 'y':
                directories[shelf_number] = []
                directories[key_shelf].remove(doc_number)
                directories[shelf_number].append(doc_number)
    else:
        print("Такой документ не существует")
```

```python
main()
```
